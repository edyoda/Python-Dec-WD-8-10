```python
soup = BeautifulSoup(response.content,"html.parser")
print(soup.prettify())
```

```python
price = soup.find("div",attrs={"class":"m-srp-card__price"})
print(price.text)
```

```python
card =  soup.find("div",attrs={"class":"m-srp-card__container"})
price = card.find("div",attrs={"class":"m-srp-card__price"})
carpet_area = card.find("div",attrs = {"class":"m-srp-card__summary__info"})
title =  card.find("a",attrs = {"class":"m-srp-card__title"})
print(title.get("href"))

response2 = requests.get(title.get("href"))
print(response2.status_code)

soup2 = BeautifulSoup(response2.content,"html.parser")
agent_name = soup2.find("p",attrs={"class":"CA_agent_name"})
if agent_name:
    agent_text = agent_name.text.replace("\n"," ")
else:
    agent_text = None
# title_text = title.text.replace("\n"," ")

# print("{} {} {}".format(title_text,price.text,carpet_area.text))
```

```python
cards =  soup.find_all("div",attrs={"class":"m-srp-card__container"})
for card in cards:
    price = card.find("div",attrs={"class":"m-srp-card__price"})
    carpet_area = card.find("div",attrs = {"class":"m-srp-card__summary__info"})
    title =  card.find("a",attrs = {"class":"m-srp-card__title"})
    title_text = title.text.replace("\n"," ")
    
    response2 = requests.get(title.get("href"))
    soup2 = BeautifulSoup(response2.content,"html.parser")
    agent_name = soup2.find("p",attrs={"class":"CA_agent_name"})
    if agent_name:
        agent_text = agent_name.text.replace("\n"," ")
    else:
        agent_text = None

    print("{} {} {} {} ".format(title_text,price.text,carpet_area.text,agent_text))
```
